# Pupil Detection

Using the CDF method mentioned in: https://arxiv.org/pdf/1202.6517.pdf

1. Facial feature detection

2. Reduce the roi

3. Use CDF method on each eye to find pupil

````python
import numpy as np
import cv2

%matplotlib inline
from matplotlib import pyplot as plt
````

## Opening and displaying

OpenCV represents RGB images as multi-dimensional NumPy arrays…but in reverse order!

This means that images are actually represented in BGR order rather than RGB!

There’s an easy fix though.

All we need to do is convert the image from BGR to RGB to display them using pyplot (otherwise, using cv2's imshow works without conversion).

````python
im = cv2.imread('../images/face.jpg')

# plt.imshow(im)
# plt.show()

im_disp = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.imshow(im_disp)
plt.show()
````

## Convert to grayscale

````python
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

# plt.axis("off")
plt.imshow(im_gray, cmap='gray')
plt.show()

# cv2.namedWindow('image', cv2.WINDOW_NORMAL)
# cv2.imshow('image',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
````

## Doing the actual face detection

````python
path = '/Users/Kepler/anaconda/pkgs/opencv3-3.1.0-py27_0/share/OpenCV/haarcascades/'

face_cascade = cv2.CascadeClassifier(path + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(path +'haarcascade_eye.xml')
````

````python
im_result = np.copy(im)
# faces = face_cascade.detectMultiScale(im_gray, 1.3, 5)

# Detect faces in the image
faces = face_cascade.detectMultiScale(
    im_gray,
    scaleFactor=1.2,
    minNeighbors=5,
#     maxSize(500, 500),
    minSize=(30, 30)
)

# the face detection function returns the (x, y, w, h) were x, y is the top left corner of the rectangle
for (x,y,w,h) in faces:
    cv2.rectangle(im_result,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = im_gray[y:y+h, x:x+w]
    roi_color = im_result[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(
        roi_gray,
        minNeighbors=5,
        minSize=(0, 0)
    )
    for (ex,ey,ew,eh) in eyes:
        print('%d, %d, %d, %d' % (ex, ey, ew, eh))
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
````

````python
im_disp = cv2.cvtColor(im_result, cv2.COLOR_BGR2RGB)
plt.imshow(im_disp)
plt.show()

# plt.imshow(im)
# plt.show()
````

## Removing false positives

````python
if eyes.shape[0] != 2:
    eyes_area = []
    for eye in eyes:
        area = eye[2] * eye[3]
        eyes_area.append(area)
    print(eyes_area)

    avg_area = sum(eyes_area) / len(eyes_area)
    
    while eyes.shape[0] != 2:
        max_diff = 0
        max_diff_index = 0
        for i in range(0, len(eyes_area)):
            diff = (avg_area - eyes_area[i])^2
            if diff > max_diff:
                max_diff = diff
                max_diff_index = i
                
#         eyes = np.delete(eyes, (i,:))
        eyes = np.delete(eyes, (i), axis=0)
````

````python
print(eyes)
print(eyes_area)
````

````python
im_result = np.copy(im)
for (x,y,w,h) in faces:
    cv2.rectangle(im_result,(x,y),(x+w,y+h),(255,0,0),2)
    roi_gray = im_gray[y:y+h, x:x+w]
    roi_color = im_result[y:y+h, x:x+w]
    for (ex,ey,ew,eh) in eyes:
        print('%d, %d, %d, %d' % (ex, ey, ew, eh))
        cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
        
im_disp = cv2.cvtColor(im_result, cv2.COLOR_BGR2RGB)
plt.imshow(im_disp)
plt.show()
````

## Cropping to get the left eye

````python
x, y = faces[0,0:2] + eyes[0,0:2]
w, h = eyes[0,2:]

left_eye = im_gray[y:y + h, x:x + w]

plt.imshow(left_eye, cmap='gray')
plt.show()
````

## CDF Method

````python
hist,bins = np.histogram(left_eye.flatten(),256,[0,256])

cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()

plt.plot(cdf_normalized, color = 'b')
plt.hist(left_eye.flatten(), 256, [0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
````

````python
real_cdf = cdf_normalized / float(cdf_normalized.max())
plt.plot(real_cdf, color = 'b')
# plt.hist(left_eye.flatten(), 256, [0,256], color = 'r')
# plt.xlim([0,256])
plt.legend(('cdf'), loc = 'upper left')
plt.show()
````

So now, to find the cdf of a certain pixel intensity, we just need to call: ```real_cdf[intensity]```.

````python
working_eye = np.copy(left_eye)

for i in range(0, working_eye.shape[0]):
    for j in range(0, working_eye.shape[1]):
        intensity = working_eye[i,j]
        cdf_value = real_cdf[intensity]
        if cdf_value < 0.05:
            working_eye[i,j] = 255
        else:
            working_eye[i,j] = 0
````

````python
plt.imshow(working_eye, cmap='gray')
plt.show()
````

## Minimum Filter

````python
def min_filter(im, radius):
    for i in range(0, im.shape[0]):
        for j in range(0, im.shape[1]):
            radius_min = 255 # the minimum intensity within the radius
            for p in range(i - radius, i + radius + 1):
                for q in range(j - radius, j + radius + 1):
                    # making sure the indices don't go out of bounds
                    row = 0 if p < 0 else p
                    col = 0 if q < 0 else q
                    row = im.shape[0]-1 if p >= im.shape[0] else p
                    col = im.shape[1]-1 if q >= im.shape[1] else q
                    
                    intensity = working_eye[row, col]
                    if intensity < radius_min:
                        radius_min = intensity
                    
            im[i,j] = radius_min
````

````python
result_eye = np.copy(working_eye)
min_filter(result_eye, 2)
````

````python
plt.imshow(result_eye, cmap='gray')
plt.show()
````

## Obtaining the Pixel of Minimum Intensity (PMI)

````python
pmi_eye = np.copy(result_eye)
pmi = 255
pmi_coords = (0, 0)
for i in range(0, pmi_eye.shape[0]):
    for j in range(0, pmi_eye.shape[1]):
        if pmi_eye[i,j] == 255:
            if left_eye[i,j] < pmi:
                pmi = left_eye[i,j]
                pmi_coords = (i, j)
````

````python
print(pmi_coords)
````

## Getting the average intensity area (11x11)

````python
ai_area = left_eye[pmi_coords[0] - 5:pmi_coords[0] + 6, pmi_coords[1] - 5:pmi_coords[1] + 6]
````

````python
ai_sum = 0
for i in range(0, ai_area.shape[0]):
    for j in range(0, ai_area.shape[1]):
        ai_sum += ai_area[i,j]
ai = float(ai_sum) / ai_area.size
````

````python
print(ai)
````

````python
final_region = left_eye[pmi_coords[0] - 7:pmi_coords[0] + 8, pmi_coords[1] - 7:pmi_coords[1] + 8]
````

````python
print(final_region.shape)
````

## Getting the pupil locations

Finding geometric center of all points in a 15x15 area around the PMI with intensities less than the AI

````python
count = 0
row_sum = 0
col_sum = 0
for i in range(pmi_coords[0] - 7, pmi_coords[0] + 8):
    for j in range(pmi_coords[1] - 7, pmi_coords[1] + 8):
        if left_eye[i, j] < ai:
            count += 1
            row_sum += i
            col_sum += j

centroid_row = float(row_sum) / count
centroid_col = float(col_sum) / count
````

````python
print('PUPIL ROW, COL:')
print('%d, %d' % (centroid_row, centroid_col))
````

http://docs.opencv.org/2.4/doc/tutorials/core/basic_geometric_drawing/basic_geometric_drawing.html

````python
final_eye = np.copy(left_eye)

pupil_row = int(centroid_row)
pupil_col = int(centroid_col)

cv2.line(final_eye, (0, pupil_row), (final_eye.shape[0] - 1, pupil_row), (0,0,0), 1)
cv2.line(final_eye, (pupil_col, 0), (pupil_col, final_eye.shape[1] - 1), (0,0,0), 1)

# wrong order of coordinates, don't use this
# cv2.line(final_eye, (pupil_row, 0), (pupil_row, final_eye.shape[0] - 1), (0,0,0), 2)
# cv2.line(final_eye, (0, pupil_col), (final_eye.shape[1] - 1, pupil_col), (0,0,0), 2)
````

````python
plt.imshow(final_eye, cmap='gray')
plt.show()
````
